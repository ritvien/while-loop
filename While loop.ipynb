```python
#Hình vuông:
import turtle
t = turtle.Turtle()
t.pencolor("red")
c = 0
while c < 4:
    t.forward(100)
    t.right(90)
    c += 1
```

```python
#rùa tập td
import turtle
import random as r
t = turtle.Turtle()
t.penup()
t.goto(-400,0)
t.pendown()
t.shape('turtle')
a = 10
while(a>0):
    t.fd(r.randint(10,50))
    t.penup()
    t.fd(r.randint(10,50))
    t.pendown()
    a-=1
```

```python
#Rùa bị quản trong nhà
import turtle
import random as r
t = turtle.Turtle()
t.penup()
t.pensize(10)
t.goto(-200,0)
t.pendown()
t.rt(90)
t.circle(200)
t.penup()
t.goto(0,0)
t.shape("turtle")
a = 1
while(a<10):
    t.rt(r.randint(1,360))
    t.fd(190)
    t.hideturtle()
    t.goto(0,0)
    t.showturtle()
    a+=1
```

```python
#Vẽ đường xoắn ốc:
import turtle
t = turtle.Turtle()
t.lt(90)
a = 10
t.speed(10)
while(a<300):
    t.circle(a,180)
    a+=15
```
